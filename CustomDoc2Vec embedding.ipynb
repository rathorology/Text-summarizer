```python
import gensim
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
import pandas as pd
```

```python
# Load File
df_train_csv = pd.read_csv('/home/rathorology/PycharmProjects/Edge/TrainData.csv', names=['ID', 'Title', 'JD', 'Category'])
df_train_csv = df_train_csv.dropna()
```

```python
txt_labels = list(df_train_csv.iloc[:,1])
txt_data = list(df_train_csv.iloc[:,2])
df_train_csv.head()
```

```python
def _clean_sent_words(sent):
    sent = str(sent).lower()
    sent = RegexpTokenizer(r'\w+').tokenize(sent)
    stop = set(stopwords.words('english'))
    sent_words = [word for word in sent if word not in stop]

    return sent_words
```

```python
class LabeledLineSentence(object):

    def __init__(self, doc_list, labels_list):

        self.labels_list = labels_list
        self.doc_list = doc_list

    def __iter__(self):

        for idx, doc in enumerate(self.doc_list):
              yield gensim.models.doc2vec.LabeledSentence(doc,    
[self.labels_list[idx]])
```

```python
txt_data = _clean_sent_words(txt_data)
it = LabeledLineSentence(txt_data, txt_labels)
```

```python
it.doc_list
```

```python
# Prepare and Train Model
model = gensim.models.Doc2Vec(size=300, min_count=0, alpha=0.025, min_alpha=0.025)
model.build_vocab(it)

for epoch in range(10000):
 print 'iteration '+str(epoch+1)
 model.train(it, total_examples=model.corpus_count, epochs=model.iter)
 model.alpha -= 0.002
 model.min_alpha = model.alpha
 model.train(it, total_examples=model.corpus_count, epochs=model.iter)

model.save('corpus_vec.model')
```
